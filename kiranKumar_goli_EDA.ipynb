```python
import pandas as pd

# Load datasets
customers = pd.read_csv(r'C:\Users\kranthi kumar goli\Downloads\Customers.csv')
products = pd.read_csv(r'C:\Users\kranthi kumar goli\Downloads\Products.csv')
transactions = pd.read_csv(r'C:\Users\kranthi kumar goli\Downloads\Transactions.csv')

# Display basic info
print(customers.info())
print(products.info())
print(transactions.info())

# Display first few rows
print(customers.head())
print(products.head())
print(transactions.head())
```

```python
# Check for missing values
print(customers.isnull().sum())
print(products.isnull().sum())
print(transactions.isnull().sum())
```

```python
# Region-wise customer distribution
region_counts = customers['Region'].value_counts()
print(region_counts)

# Most popular product categories
popular_categories = products['Category'].value_counts()
print(popular_categories)

# Total revenue generated
total_revenue = transactions['TotalValue'].sum()
print(f"Total Revenue: ${total_revenue}")

# Average product price
avg_price = products['Price'].mean()
print(f"Average Product Price: ${avg_price:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for regions
sns.countplot(data=customers, x='Region')
plt.title('Customer Distribution by Region')
plt.show()

# Top product categories
products['Category'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Top Product Categories')
plt.show()
```
